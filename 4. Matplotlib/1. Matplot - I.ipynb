<b>plt.figure()</b>

- If a figure with that identifier already exists, this figure is made active and returned

- If there is no figure with the identifier or num is not given, a new figure is created, made active and returned.

<b>plt.show()</b>

- Display all open figures.

- In non-interactive mode, block defaults to True. All figures will display and show will not return until all windows are closed. If there are no figures, return immediately.

```python
plt.figure()
plt.plot(X, X**2, label="square")  # label -> To give a label to the curve
plt.plot(X, X**3, label="cube")
plt.legend()
plt.ylim(-100, 100)
plt.xlim(-7.5, 7.5)
plt.show()
```

<b> plt.legend()</b>

- legend is used to display the labels of the curves. Without writing plt.legend(), labels will not be displayed

```python
plt.figure()
plt.plot(X, X**2, label="square")
plt.plot(X, X**3, label="cube")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My plot")
plt.legend()
plt.ylim(-100, 100)
plt.xlim(-7.5, 7.5)
plt.show()
```

<b>plt.xlabel()</b> and <b>plt.ylabel</b>

- Used to give labels to the X and Y axis of the graph

```python
plt.figure()
plt.plot(X, X**2, label="square")
plt.plot(X, X**3, label="cube")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My plot")
plt.legend()
plt.ylim(-100, 100)
plt.xlim(-7.5, 7.5)

plt.plot(X, X**2, label="square")
plt.plot(X, X**3, label="cube")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My plot")
plt.legend()
plt.ylim(-100, 100)
plt.xlim(-7.5, 7.5)
plt.show()
```

```python
# We can see without using plt.show() or plt.figure() in between the two blocks of code, we are getting overlapping graphs.
# However, if we just give plt.figure() in between the two blocks, we would get two seperate hgraphs
```

```python
plt.figure()
plt.plot(X, X**2, label="square")
plt.plot(X, X**3, label="cube")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My plot")
plt.legend()
plt.ylim(-100, 100)
plt.xlim(-7.5, 7.5)

plt.figure()
plt.plot(X, X**2, label="square")
plt.plot(X, X**3, label="cube")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My plot")
plt.legend()
plt.ylim(-100, 100)
plt.xlim(-7.5, 7.5)
plt.show()
```

### Graph of sin, cos,tan

```python
X = np.linspace(-np.pi, np.pi, 100)
```

```python
y = np.sin(X)
```

```python
plt.plot(X, y)
```

```python
y = np.cos(X)
```

```python
plt.plot(X, y)
```

```python
y = np.tan(X)
```

```python
plt.figure()
plt.plot(X, np.sin(X), label="sin")
plt.plot(X, np.cos(X), label="cos")
plt.plot(X, np.tan(X), label="tan")
plt.legend()
plt.ylim(-4, 4)
plt.show()
```

### Plotting a circle

```python
theta = np.linspace(-np.pi, np.pi, 100)
rad=10
```

```python
X = rad * np.cos(theta)
y = rad * np.sin(theta)
```

```python
plt.plot(X, y)
```

```python
# It created an oval like shape. it happens because our axes are scaled.
# To fix this, what we can do is that in plt.figure(), we can specify the size of the figure to scale to in inches.

# By default, figsize is (6.4, 4.8)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y)
```

### Pie Chart

```python
plt.pie([10, 20, 30])
```

```python
# Starts from degree 0 and displays the items on the basis of fraction of total area
```

```python
plt.pie([10, 20, 30], shadow = True)  # shadow is added
```

```python
out = plt.pie([10, 20, 30], shadow = True, explode=[.1, 0, 0]) # Seperates an item of the pie chart from the rest 
```

```python
out = plt.pie([10, 20, 30], shadow = True, explode=[.1, 0, 0], startangle=30)   # Starts placing the items from 30 degrees
```

```python
x       = [10, 20, 30]
explode = [.1, 0, 0]
labels  = ["Apple", "Google", "Facebook"]
colors  = ["black", "red", "blue"]
out = plt.pie(x, shadow = True, explode=explode, startangle=30, colors = colors, labels=labels)
```
