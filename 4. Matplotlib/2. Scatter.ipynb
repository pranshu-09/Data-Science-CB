### Scatter

```python
x = np.linspace(-20, 20, 10)
```

```python
y = x ** 2 
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, marker="^")
```

```python
mean = [10, 20]
covar = [[4,0], [0,5]]
```

```python
data = np.random.multivariate_normal(mean, covar, size=1000)
```

```python
data.shape
```

```python
f1 = data[:,0]
```

```python
f2 = data[:, 1]
```

```python
plt.scatter(f1, f2)
```

```python
f1.var()
```

```python
f2.var()
```

```python
mean1 = [10, 20]
covar1 = [[4,0], [0,5]]

data1 = np.random.multivariate_normal(mean1, covar1, size=100)
```

```python
mean2 = [40, 45]
covar2 = [[4,0], [0,5]]

data2 = np.random.multivariate_normal(mean2, covar2, size=100)
```

```python
plt.scatter(data1[:,0], data1[:,1])
plt.scatter(data2[:,0], data2[:,1])
```
