```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("RGB_Matte.png")
```

```python
img
```

```python
small = img.resize((400, 300))
```

```python
small
```

```python
type(small)
```

```python
# We need to convert this PIL.Image.Image into a numpy array
```

```python
panda = np.array(small)
```

```python
type(panda)
```

```python
panda.shape
```

```python
red = panda[:, :, 0]
green = panda[:, :, 1]
blue = panda[:, :, 2]
```

```python
plt.imshow(red, cmap="gray")
```

```python
# We can see we were able to get the part of the image that was red
```

```python
plt.imshow(green, cmap="gray")
```

```python
plt.imshow(blue, cmap="gray")
```

```python
panda[:, :, 1:] = 0
```

```python
plt.imshow(panda)
```
